```python
def roc(close, period = 12):
    """
    Rate of Change

    params:

    close: close price
    period: lookback period (default: 12)

    close: close price
    rsi_param: parameter used for calculation of RSI applied to the streak (default: 2)

    """
    #close = close.rank(pct=True)

    return pd.Series((close.diff(period) / close.shift(period)) * 100, name="ROC")
```

```python
def msv(cls, rsi_param = 2):
    """
    Market Streak Value

    params:

    close: close price
    rsi_param: parameter used for calculation of RSI applied to the streak (default: 2)

    """
    #Price streak
    streak = cls.copy()
    shifted = streak.shift()

    streak["streak_start"] = np.where((shifted < streak), 1, np.where(shifted > streak, -1, 0))

    streak['streak_id'] = np.where((streak["streak_start"] == 1), 1, np.where((streak["streak_start"] == -1), -1, 0))


    streak['streak_counter'] = streak.groupby('streak_id').cumcount() + 1

    for i in range(len(streak.streak_id)):
        if streak.streak_id[i] == 0:
            streak.streak_counter[i] = 0
        elif streak.streak_id[i] == streak.streak_id[i-1]:
            streak.streak_counter[i] = streak.streak_id[i] + streak.streak_counter[i-1]
        else:
            streak.streak_counter[i] = streak.streak_id[i]



    #Calculation of MSV with 2 period RSI
    msv_ = vbt.RSI.run(streak['streak_counter'], window=rsi_param).rsi



    return msv_
```

```python
def connors(close, rsi_window = 3,  roc_period = 10, msv_window = 2):
    """
    Connors RSI
    https://www.quantifiedstrategies.com/connors-rsi/

    params:

    close: close price
    entry: entry level of Connor's RSI (default: 5)
    exit: exit level of Connor's RSI (default: 95)
    roc_period: lookback input parameter used for calculation of ROC (default: 100)
    msv_window: lookback input for RSI calculation of Market Streak Value (default: 2)
    rsi_window: window used for calculation of RSI for connors RSI (default: 3)

    """

    # 3-period RSI
    rsi_ = vbt.RSI.run(close["close"], window=rsi_window).rsi.squeeze()


    #Market Streak Value
    msv_ = msv(cls=close, rsi_param=msv_window)


    #ROC

    roc_ = roc(close['close'], period=roc_period)


    #Connors RSI

    c_rsi = ((rsi_ + msv_ + roc_)/3)


    trend = np.where(c_rsi > 60, -1, 0)
    trend = np.where((c_rsi  < 10), 1, trend)
    trend = pd.Series(data=trend, index = c_rsi.index)

    return trend
```

```python
ind = vbt.IndicatorFactory(
    class_name="Connors RSI",
    short_name="CRSI",
    input_names=["close"],
    param_names=["rsi_window", "roc_period", "msv_window"],
    output_names=["value"]
).from_apply_func(connors, rsi_window=3, roc_period=100, msv_window=2, keep_pd = True)


res = ind.run(data['close'], rsi_window=3, roc_period=100, msv_window=2)

entry = res.value == 1.0
exit = res.value == -1.0


pf = vbt.Portfolio.from_signals(data['close'], entry, exit)
print(pf.total_return())
```
